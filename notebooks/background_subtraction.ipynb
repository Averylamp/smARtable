```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
first = cv2.imread("/Users/ethanweber/Documents/github_personal/smARtable/frame_1.jpg")
```

```python
second = cv2.imread("/Users/ethanweber/Documents/github_personal/smARtable/frame_6.jpg")
```

```python
plt.imshow(first)
```

```python
f_g = cv2.cvtColor(first, cv2.COLOR_BGR2GRAY)
```

```python
s_g = cv2.cvtColor(second, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(f_g, cmap='gray')
```

```python
plt.imshow(s_g, cmap='gray')
```

```python
value = np.absolute(np.subtract(f_g, s_g))
```

```python
plt.imshow(value, cmap='gray')
```

```python
clipped = np.clip(value, 1, 255)
```

```python
plt.imshow(clipped, cmap='gray')
```
